## matplotlib基本语法

```python
from matplotlib import pyplot as plt
import random
x = range(2,26,2)
y = [11, 23, 11.3, 35, 45, 65, 35, 27, 76, 8, 11, 45]
plt.plot(x,y)
plt.show()
```

## 描述信息

保存图片

```python
fig = plt.figure(figsize=(10,5),dpi=100)
x = range(2, 26, 2)
y = [11, 23, 11.3, 35, 45, 65, 35, 27, 76, 8, 11, 45]
plt.plot(x, y)
plt.savefig("1.svg")
# 保存图片为svg格式，放大不会有模糊
plt.show
```

设置刻度

```python
x = range(2, 26, 2)
y = [11, 23, 11.3, 35, 45, 65, 35, 27, 76, 8, 11, 45]
plt.plot(x, y)
plt.xticks(x)
plt.yticks(range(min(y),max(y)+1)[::5])
plt.show()
```

设置xlabel

```python
x = range(0, 60)
y = [random.randint(10, 40) for i in range(60)]
plt.plot(x, y)
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x,_xtick_labels,rotation=50)
plt.show()
```

引入中文字体

```python
import matplotlib
font = {"family":"MicroSoft YaHei",
       "weight":"bold",
       "size":"12"}
matplotlib.rc("font", **font)
x = range(0,60)
y = [random.randint(10,40) for i in range(60)]
plt.plot(x,y)
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
plt.show()
```

加入轴名称

```python
font = {"family":"MicroSoft YaHei",
       "weight":"bold",
       "size":"12"}
matplotlib.rc("font", **font)
x = range(0,60)
y = [random.randint(10,40) for i in range(60)]
plt.plot(x,y)
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("每分钟气温变化")
plt.grid(alpha=0.4)
plt.show()
```

设置线条颜色 格式 粗细 透明度

```python
font = {"family":"MicroSoft YaHei",
       "weight":"bold",
       "size":"12"}
matplotlib.rc("font", **font)
x = range(0,60)
y = [random.randint(10,40) for i in range(60)]
plt.plot(x,y,
        color="red",
         linestyle="--",
         linewidth=2,
         alpha=0.5
        )
_x = x[::5]
_xtick_labels = ["{}分钟".format(i) for i in _x]
plt.xticks(_x, _xtick_labels, rotation=50)
plt.xlabel("时间")
plt.ylabel("温度")
plt.title("每分钟气温变化")
plt.grid(alpha=0.4)
plt.show()
```
